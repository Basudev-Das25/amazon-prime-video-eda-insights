# **Understanding of Variables**

###**Dataset Description**

The dataset is related to movies and TV shows available on Amazon Prime Video. It's formed by merging two datasets: titles.csv and credits.csv.

**title.csv** dataset contains information about movies and TV shows available on Amazon Prime Video. Each row represents a single title.

**credits.csv** dataset contains information about the people involved in the production of the titles (actors, directors, etc.). Each row represents a person's involvement in a specific title.

###**Dataset Columns**

**id**: Unique identifier for each title (movie or TV show)

**title**: Title of the movie or TV show

**type**: Whether it's a 'SHOW' (TV show) or a 'MOVIE'

**description**: A brief description of the title

**release_year**: The year the title was released

**age_certification**: Age rating or certification (e.g., PG-13, TV-MA)

**runtime**: Duration of the movie or TV show in minutes

**genres**: List of genres associated with the title (e.g., ['comedy', 'drama'])

**production_countries**: List of countries where the title was produced

**seasons**: Number of seasons (for TV shows only)

**imdb_id**: IMDb identifier for the title

**imdb_score**: IMDb rating of the title

**imdb_votes**: Number of votes the title received on IMDb

**tmdb_popularity**: Popularity score from TMDb (The Movie Database)

**tmdb_score**: Rating score from TMDb

**person_id**: Unique identifier for a person involved in the production (actor, director, etc.)

**name**: Name of the person (actor, director, etc.)

**character**: The character played by the person (for actors)

**role**: The role of the person in the production (e.g., 'ACTOR', 'DIRECTOR')

#**Data Wrangling**

###**Merging of Dataset**

```python
#merge 2 dataset on common column id
data = pd.merge(data1,data2,on='id')
```

###**Missing values/Null values**

```python
#check for null values
data.isnull().sum()
```

We have null values for **description column**. Let's drop the null values from **description column**.

```python
#drop null values in description column
data.dropna(subset = ['description'], inplace = True)
```

We have null valuess for age_certification column.

```python
#Mode of age_certification
data['age_certification'].mode()[0]
```

```python
#Replace all null values of age_certification with mode
data['age_certification'].fillna(data['age_certification'].mode()[0], inplace = True)
```

```python
#Null value check for the merged dataset
data.isnull().sum()
```

The 'seasons' column has null values because it's only applicable to TV shows, Movies don't have seasons so these entries would naturally be null. Instead of removing these rows,  replace them wih 0.

```python
#replacing null values in seasons with 0
data['seasons'].fillna(0, inplace = True)
```

```python
#We check the dataset again
data.isnull().sum()
```

Since 'imdb_id' is unique for Movies and TV shows, we will drop all null values in 'imdb_id' column.

```python
#droping all null values in imdb_id column
data.dropna(subset=['imdb_id'], inplace=True)
```

As we know 'imdb_score' represents the rating of a Movie or TV shows, removing null values in 'imdb_score' might result to data loss, so it's better to replace them with mean value of the column.

```python
#replacing the null values in imdb_score with mean value of the column
data['imdb_score'].fillna(round(data['imdb_score'].mean(),1), inplace=True)
```

'imdb_votes' represent the number of votes a Movie or TV show attain. Missing values in 'imdb_votes' may indicate that either the Movie or TV show is new or have not garnered much attention. So, it's better to replace the null value wih 0

```python
#replacing null values in imdb_votes with 0
data['imdb_votes'].fillna(0, inplace=True)
```

We check the dataset again for null values

```python
data.isnull().sum()
```

Similar to 'imdb_score', 'tmdb_popularity' and 'tmdb_score' will also have it's null values replaced with the mean value of the column.

```python
#replacing null values in 'tmdb_popularity' column with mean
data['tmdb_popularity'].fillna(round(data['tmdb_popularity'].mean(),2), inplace=True)
```

```python
#replacing null values in 'tmdb_scores' column with mean
data['tmdb_score'].fillna(round(data['tmdb_score'].mean(),1), inplace=True)
```

Missing values in he column 'characters' column indicate that the information about the character played by a person in movie or TV show is not available in the dataset. So we replace the null values by 'unknown'

```python
#replacing the null values in 'characters' with unknown
data['character'].fillna('unknown', inplace=True)
```

We make a final check for null values.

```python
data.isnull().sum()
```

###**Multivariate Analysis**

####**Country vs Genre**

```python
# Copy dataset
df = data.copy()

# --- Step 1: Clean columns ---
# Convert string lists to actual lists
df['production_countries'] = df['production_countries'].astype(str)
df['genres'] = df['genres'].astype(str)

# Keep only first country
df['production_countries'] = df['production_countries'].str.strip("[]").str.split(',').str[0]

# Split genres
df['genres'] = df['genres'].str.strip("[]").str.split(',').str[0]

# Remove quotes/spaces
df['production_countries'] = df['production_countries'].str.replace("'", "").str.strip()
df['genres'] = df['genres'].str.replace("'", "").str.strip()

# --- Step 2: Create pivot table ---
pivot = pd.crosstab(df['production_countries'], df['genres'])

# Select top countries (by total content)
top_countries = pivot.sum(axis=1).sort_values(ascending=False).head(10).index
pivot_top = pivot.loc[top_countries]

# --- Step 3: Plot ---
plt.figure(figsize=(14, 8))

sns.heatmap(
    pivot_top,
    cmap='viridis',
    annot=True,
    fmt='d',
    linewidths=0.5,
    cbar_kws={'label': 'Count'}
)

plt.title("Top Countries vs Genre Distribution", fontsize=14)
plt.xlabel("Genres")
plt.ylabel("Production Countries")

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

Why this chart?

  * A heatmap is ideal for comparing two categorical variables with a numerical value
  * Helps quickly identify high-volume genre-country combination

Insight

  * United States dominates across all genres, especially Drama, Comedy, Thriller, Romance
  * Drama is the most dominant genre globally followerd by comedy and thriller

Business Impact

  * Helps identify which genre performs well in specific regions, which is useful for reginal content strategy
  * Enables better localization and recommendation system based on region-genre patterns
